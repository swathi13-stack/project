Now replace the null values with the mean or median  or zeroes of the corresponding columns.
Here we used median as for each country the properties of columns differs and by using median it won't  much effect the predicting column (average temperature in degrees).

```python
df['Crop Residues']=df['Crop Residues'].isna().fillna(0)
df['Savanna fires']=df.groupby('Area')['Savanna fires'].fillna(df['Savanna fires'].median())
df['Forest fires']=df.groupby('Area')['Forest fires'].fillna(df['Forest fires'].median())
df['Forestland']=df.groupby('Area')['Forestland'].fillna(df['Forestland'].median())
df['Net Forest conversion']=df.groupby('Area')['Net Forest conversion'].fillna(df['Net Forest conversion'].median())
df['Food Household Consumption']=df.groupby('Area')['Food Household Consumption'].fillna(df['Food Household Consumption'].median())
df['IPPU']=df.groupby('Area')['IPPU'].fillna(df['IPPU'].median())
df['Manure applied to Soils']=df.groupby('Area')['Manure applied to Soils'].fillna(df['Manure applied to Soils'].median())
df['Manure Management']=df.groupby('Area')['Manure Management'].fillna(df['Manure Management'].median())
df['Fires in humid tropical forests']=df.groupby('Area')['Fires in humid tropical forests'].fillna(df['Fires in humid tropical forests'].median())
df['On-farm energy use']=df.groupby('Area')['On-farm energy use'].fillna(df['On-farm energy use'].median())
```
